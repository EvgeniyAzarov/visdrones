```python
import matplotlib.pyplot as plt
import matplotlib.patches as patches
from PIL import Image 
import pandas as pd
```

```python
img = Image.open("/home/yaz/mlc/baseline/datasets/VisDrone/VisDrone2019-DET-train/images/0000002_00005_d_0000014.jpg")
```

```python
width, height = img.size
```

```python
annot_file = "/home/yaz/mlc/baseline/datasets/VisDrone/VisDrone2019-DET-train/labels/0000002_00005_d_0000014.txt" 
annot = pd.read_csv(annot_file, names=["class", "x", "y", "w", "h"], sep="\s+")
annot['x'] = (annot['x'] * width).astype(int)
annot['w'] = (annot['w'] * width).astype(int) 
annot['y'] *= height
annot['h'] *= height
annot
```

```python
fig, ax = plt.subplots(figsize=(20, 6))
ax.imshow(img)

for i, (cls, x, y, w, h) in annot.iterrows():
    rect = patches.Rectangle((x, y), w, h, facecolor="none", linewidth=1, edgecolor='r') 
    ax.add_patch(rect)

plt.show()
```
